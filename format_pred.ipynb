```python
pred_path="/home/u131168/mh_shell/data/custom_pred/full_pred7.csv"

formatted_sub_path="/home/u131168/mh_shell/data/shell_sub_custom7.csv"

full_data_path="/home/u131168/mh_shell/data/f_data.csv"

full_data=pd.read_csv(full_data_path)

pred_data=pd.read_csv(pred_path,sep="\$\$", header=None)
pred_data.columns=['index','answer']
# pred_data=pred_data.iloc[6000:7000]
pred_data.fillna("", inplace = True)
# unknown="_."  , 
pred_data.isna().sum()
pred_data=pred_data.drop_duplicates()

# filtering pred results
pred_data['answer']=pred_data['answer'].apply(lambda x: x if 'ț' not in x else str(x).replace('ț',''))
# pred_data['answer']=pred_data['answer'].apply(lambda x: x if '>' not in x else str(x).replace('>','<>'))


pred_data_c=pred_data.copy()

# keeping only predicted_vals
pred_data=pred_data[full_data['output'].isna()]

pred_data_c.loc[full_data['output'].notnull(),'answer']=full_data[full_data['output'].notnull()]['output']


pred_data['answer']=pred_data['answer'].apply(lambda x: "" if str(x).strip()=="none" else x )
pred_data_c['answer']=pred_data_c['answer'].apply(lambda x: "" if str(x).strip()=="none" else x )


pred_data
# pred_data[full_data['output'].notnull()][:20]
# pred_data[pred_data.isna().any(axis=1)]
```
